```python
import MySQLdb
import cv2
import requests
import urllib.parse
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import datetime



connection = MySQLdb.connect(host = "127.0.0.1" , user = "root" , password = "zxcv2587", db = 'test' , charset = 'utf8') 
cursor = connection.cursor()
cursor.execute('TRUNCATE TABLE `test`.`gasyas_2019`')










def get_data(url):
    
    req = requests.get(url)
    bs = BeautifulSoup(req.text , 'lxml')
    
    temp = bs.select('.ui_anchor_container')
    
    
    
    
    start_date = []
    end_date = []
    name_jp = []
    ori_url = []
    image = []
    
    
    for i in temp:
        if len(i.select('a')) != 0:
            temp = i.select('a')[0]
            if len(temp.text) != 0:
                line = temp.text
                if(line.find('～') == -1):
                    continue
                
                
                
                start_date.append(line[:line.find('～')])
                line = line[line.find('～') + 1:]
                end_date.append(line[:10])
                name_jp.append(line[10:])
                
                a = i.select('a')[0]
                
                ori_url.append(urllib.parse.unquote(a['href']))
    
    
    
    
    
    
    
    img = bs.select('.t-line-img')
    for i in img:
        line = i.select('td')[0]
        line = line.select('img')[0]
        image.append(line['data-original'])
            
            
    
    count = 300
    for i in range(0 , len(name_jp)):
        count = count + 1
        print(name_jp[i])
        print(start_date[i])
        print(end_date[i])
        print(image[i])
        print(ori_url[i])
        
        
        
        q = cursor.execute('INSERT into `test`.`gasyas_2019`(`id` , `name_jp` , `name_tw` , `start_date` , `end_date` , `image` , `ori_url`) VALUES(%s , %s , %s , %s , %s , %s , %s)' ,
        (count , name_jp[i] , name_jp[i] , start_date[i] , end_date[i] , image[i] , ori_url[i]))
        
        
        
    


homepage_2019 = 'https://imasml-theater-wiki.gamerch.com/%E9%81%8E%E5%8E%BB%E3%81%AE%E3%82%AC%E3%82%B7%E3%83%A3%202019%E5%B9%B4'

get_data(homepage_2019)




connection.commit()
cursor.close()
connection.close()
print('papara')
```
