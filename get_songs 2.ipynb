```python
import MySQLdb
import cv2
import requests
import urllib.parse
from bs4 import BeautifulSoup
from urllib.parse import urljoin



connection = MySQLdb.connect(host = "127.0.0.1" , user = "root" , password = "zxcv2587", db = 'test' , charset = 'utf8') 
cursor = connection.cursor()
#cursor.execute('TRUNCATE TABLE `test`.`songs`')



def get_data(url):
    print(url)
    req = requests.get(url)
    bs = BeautifulSoup(req.text , 'lxml')
    part = bs.select('.ui_wikidb_middle_area')[0]
    line = part.select('p')
    count = 0
    
    value = []
    
    for i in line:
        count = count + 1
        if count % 2 == 0:
            
            continue
        temp = i.text
        pos = temp.find(')') + 2
        #temp[pos:])
        number = int(temp[pos:])
        value.append(number)
        
    
    name = bs.select('.ui_page_match')[0].text
    img = bs.select('.lazy')[0]
    #img['src']
    
    temp = bs.select('.ui_wikidb_top_area')[0]
    part = temp.select('p')
    song_type = part[0].text[5:]
    #print(song_type)
    BPM = part[1].text[4:]
    #print(BPM)
    length = part[2].text[3:]
    length = '00:' + length
    #print(length)
    the_date = part[3].text[4:]
    #print(date)
    
    temp = bs.select('.ui_wiki_db_bottom_wrapper')[0]
    line = temp.select('p')
    
    line = line[1].text[5:]

    
    
    
    q = cursor.execute('''
    INSERT INTO `test`.`songs` (`name` ,
    `2M.LV` , `2M.notes` , `2M+.LV` , `2M+.notes` ,
    `4M.LV` , `4M.notes` , `6M.LV` , `6M.notes` ,
    `MM.LV` , `MM.notes` , 
    `BPM` , `length` , `type` ,
    `date` , 
    `get_song`, 
    `image`)
    VALUES (%s , 
    %s , %s , %s , %s,
    %s , %s , %s , %s,
    %s , %s ,
    %s , %s , %s , %s , %s , %s)
    ''' ,
    (name ,value[0] , value[1] , value[2] , value[3] ,value[4] , value[5] , value[6] , value[7] ,value[8] , value[9] ,
    BPM , length , song_type , the_date , line ,img['src']))
    
    

    
    
def get_url(hp):
    
    
    req = requests.get(hp)
    bs = BeautifulSoup(req.text , 'lxml')
    table = bs.select('table')[0]
    body = table.select('tbody')[0]
    lines = body.select('td')
    for line in lines:
        if line['data-col'] == '0':
            part = line.select('a')[0]
            url = part['href']
            get_data(urljoin(hp , url))
    
    
    
homepage = 'https://imasml-theater-wiki.gamerch.com/%E3%83%A9%E3%82%A4%E3%83%96%E3%82%B3%E3%83%BC%E3%82%B9%E5%88%A5%E6%A5%BD%E6%9B%B2%E4%B8%80%E8%A6%A7'
get_url(homepage)



connection.commit()
cursor.close()
connection.close()
```
