```python
import MySQLdb
import cv2
import requests
import urllib.parse
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import datetime



connection = MySQLdb.connect(host = "127.0.0.1" , user = "root" , password = "zxcv2587", db = 'test' , charset = 'utf8') 
cursor = connection.cursor()
cursor.execute('TRUNCATE TABLE `test`.`gasya`')





def take_url(url):
    
    req = requests.get(url)
    bs = BeautifulSoup(req.text , 'lxml')
    temp = bs.select('meta')
    for i in temp:
        line = str(i)
        if line.find('keywords') != -1:
            line = str(i['content'])
            return  line[line.rfind(',') + 1 :]




def get_data(url):
    
    req = requests.get(url)
    bs = BeautifulSoup(req.text , 'lxml')
    
    string = str(bs)
    pos_A = string.find('id="content_1_1"')
    pos_B = string.find('style="text-align:right"')
    part = string[pos_A : pos_B -5]
    pos_A = part.find('</div>')
    
    
    
    bs = BeautifulSoup(part[pos_A + 7:] , 'lxml')
    
    
    temp = bs.select('table')
    count = 0
    image_list = []
    for i in temp:
        count = count + 1
        if count % 2 == 1:
            line = i.select('img')[0]
            image_list.append(line['data-original'])
            
    
    
    
    
    
    
    
    
    
    temp = bs.select('.ui_anchor_container')
    gasya = []
    for i in temp:
        gasya.append(i.text)
        

        
    count = 0
    temp = bs.select('img')
    part = []
    
    cards = ''
    for i in temp:
        try:
            cards = cards + i['alt'] + '\n'
        except:
            cards = cards[:-1]
            part.append(cards)
            cards = ''
    
    
    
    url_list = []
    req = requests.get(url)
    bs = BeautifulSoup(req.text , 'lxml')
    
    sth = bs.select('.ui_h-large')
    
    for i in sth:
        line = i.select('a')
        if len(line) != 0:
            a = line[0]['href']
            url_list.append(a)
            
    
    
    
    
    
    title = ''
    
    for i in range(0 , len(gasya)):
        line = gasya[i]
        pos = line.find('～')
        start_date = line[:pos]
        end_date = line[pos + 1: pos + 11]
        
        if line[pos +11] == '【':
            title = line[pos + 11 :]
        else:
            title = line[pos + 12: ]
            
        content = part[i]
        a = take_url(url_list[i])
        print(a)
        hp = 'https://imasml-theater-wiki.gamerch.com/'
        
        
        
        
        
        q = cursor.execute('INSERT INTO `test`.`gasya` (`title` , `title_tw` , `cards` , `start_date` , `end_date` , `image` , `url`) VALUES (%s , %s , %s , %s , %s , %s , %s)' , 
                           (title , title , content , start_date, end_date , image_list[i] , hp + a))
    


homepage_2020 = 'https://imasml-theater-wiki.gamerch.com/%E9%81%8E%E5%8E%BB%E3%81%AE%E3%82%AC%E3%82%B7%E3%83%A3'
homepage_2019 = 'https://imasml-theater-wiki.gamerch.com/%E9%81%8E%E5%8E%BB%E3%81%AE%E3%82%AC%E3%82%B7%E3%83%A3%202019%E5%B9%B4'
homepage_2018 = 'https://imasml-theater-wiki.gamerch.com/%E9%81%8E%E5%8E%BB%E3%81%AE%E3%82%AC%E3%82%B7%E3%83%A3%EF%BC%882018.01.15%EF%BD%9E2018.12.31%EF%BC%89'
homepage_2017 = 'https://imasml-theater-wiki.gamerch.com/%E9%81%8E%E5%8E%BB%E3%81%AE%E3%82%AC%E3%82%B7%E3%83%A3%EF%BC%882017.06.29%EF%BD%9E2017.12.30%EF%BC%89'
get_data(homepage_2020)
get_data(homepage_2019)
get_data(homepage_2018)
get_data(homepage_2017)




connection.commit()
cursor.close()
connection.close()
print('papara')
```
