```python
import csv

f = open('code.txt' , 'w' ,  encoding= 'utf-8')

with open('songs.csv' , newline='' , encoding = 'utf-8') as csvfile:
    
    rows = csv.reader(csvfile)
    part = []
    flag = 1
    for row in rows:
        if flag:
            flag= 0
            continue;
        part.append('$song = new App\Song();')
        part.append('$song->name = \'' + row[0] +'\';')
        part.append('$song->EZ_LV = ' + row[1] + ';')
        part.append('$song->EZ_notes = ' + row[2] + ';')
        part.append('$song->NM_LV = ' + row[3] + ';')
        part.append('$song->NM_notes = ' + row[4] + ';')
        part.append('$song->HR2_LV = ' + row[5] + ';')
        part.append('$song->HR2_notes = ' + row[6] + ';')
        part.append('$song->HR_LV = ' + row[7] + ';')
        part.append('$song->HR_notes = ' + row[8] + ';')
        part.append('$song->EX_LV = ' + row[9] + ';')
        part.append('$song->EX_notes = ' + row[10] + ';')
        part.append('$song->BPM = \'' +row[11]+'\';')
        part.append('$song->length = \''+row[12]+'\';')
        part.append('$song->type = \''+row[13]+'\';')
        part.append('$song->date = \''+row[14]+'\';')
        part.append('$song->image = \''+row[15]+'\';')
        part.append('$song->get_song = \'' +row[16] + '\';')
        part.append('$song->save();')
        
        for line in part:
            f.write(line)
            f.write('\n')
        print(' ')
        

f.close()
#$song = new App\Song;
```
